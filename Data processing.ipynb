```python
# Imports
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from scipy import stats
import pandas as pd
import numpy as np
import os
```

```python
dataFrame = pd.read_csv('data/data.csv', sep=';', decimal=",")
```

```python
dataFrame.describe()
```

```python
# print(dataFrame.dtypes)

dataFrame['duration'] = dataFrame['e'] - dataFrame['b']
dataFrame = dataFrame.drop(['b', 'e'], axis=1)
dataFrame = dataFrame[['duration'] + dataFrame.columns.tolist()[:-1]]
dataFrame.describe()
```

```python
#Commenting a line results in not normalizing that attribute
normalized_columns = ['AC', 
                      'FM', 
                      'UC', 
                      'DL', 
                      'DS', 
                      'DP', 
                      'DR', 
                      'LB', 
                      'AC.1',
                      'FM.1', 
                      'UC.1', 
                      'DL.1', 
                      'DS.1', 
                      'DP.1', 
                      'ASTV', 
                      'MSTV', 
                      'ALTV', 
                      'MLTV',
                      'Width', 
                      'Min', 
                      'Max', 
                      'Nmax', 
                      'Nzeros', 
                      'Mode', 
                      'Mean', 
                      'Median',
                      'Variance', 
                      'Tendency' 
                      #'A', 
                      #'B', 
                      #'C', 
                      #'D', 
                      #'E', 
                      #'AD', 
                      #'DE', 
                      #'LD', 
                      #'FS',
                      #'SUSP'
                     ]

# Normalize the data
scaler = MinMaxScaler()
dataFrame[normalized_columns] = pd.DataFrame(data=scaler.fit_transform(dataFrame[normalized_columns]))
```
